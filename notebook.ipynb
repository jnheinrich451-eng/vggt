```python
from pathlib import Path

import torch
from vggt.models.vggt import VGGT
from vggt.utils.load_fn import load_and_preprocess_images
from vggt.utils.pose_enc import pose_encoding_to_extri_intri

device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.bfloat16 if torch.cuda.is_available() and torch.cuda.get_device_capability()[0] >= 8 else torch.float16

# Load VGGT
model = VGGT.from_pretrained("facebook/VGGT-1B").to(device).eval()

# Use sampled frames from your video in chronological frame order.
image_names = [str(path) for path in sorted(Path("frames").glob("*.png"))]

images = load_and_preprocess_images(image_names).to(device)  # [S, 3, H, W]
images = images[None]  # [B, S, 3, H, W]

with torch.no_grad():
    with torch.cuda.amp.autocast(dtype=dtype):
        aggregated_tokens_list, ps_idx = model.aggregator(images)

    # Camera-only prediction
    pose_enc = model.camera_head(aggregated_tokens_list)[-1]

    # Convert compact camera encoding to matrices
    extrinsic, intrinsic = pose_encoding_to_extri_intri(
        pose_enc,
        images.shape[-2:]
    )

```

```python
print("extrinsic:", extrinsic.shape)  # [B, S, 3, 4]
print("intrinsic:", intrinsic.shape)  # [B, S, 3, 3]
print("extrinsic:", extrinsic[0, :5,:,:])  # [5, 3, 4]
print("intrinsic:", intrinsic[0, :5,:,:])  # [5, 3, 3]
```

```python
import torch

def w2c_3x4_to_c2w_4x4(extrinsic):
    # extrinsic: [B, F, 3, 4], VGGT world-to-camera
    B, F = extrinsic.shape[:2]
    bottom = torch.zeros(B, F, 1, 4, device=extrinsic.device, dtype=extrinsic.dtype)
    bottom[..., 0, 3] = 1
    w2c = torch.cat([extrinsic, bottom], dim=-2)
    return torch.linalg.inv(w2c)

def make_relative_c2w(c2w):
    # CameraCtrl inference normalizes first frame to identity.
    first_w2c = torch.linalg.inv(c2w[:, :1])
    return first_w2c @ c2w

def intrinsics_to_k4(intrinsic, source_hw, target_hw):
    # CameraCtrl wants [fx, fy, cx, cy] in pixels for its target H/W.
    src_h, src_w = source_hw
    tgt_h, tgt_w = target_hw

    sx = tgt_w / src_w
    sy = tgt_h / src_h

    fx = intrinsic[..., 0, 0] * sx
    fy = intrinsic[..., 1, 1] * sy
    cx = intrinsic[..., 0, 2] * sx
    cy = intrinsic[..., 1, 2] * sy

    return torch.stack([fx, fy, cx, cy], dim=-1)

def plucker_from_cameras(K, c2w, H, W):
    # K: [B, F, 4] = fx, fy, cx, cy
    # c2w: [B, F, 4, 4]
    B, F = K.shape[:2]
    device = c2w.device
    dtype = c2w.dtype

    j, i = torch.meshgrid(
        torch.arange(H, device=device, dtype=dtype),
        torch.arange(W, device=device, dtype=dtype),
        indexing="ij",
    )

    i = i.reshape(1, 1, H * W).expand(B, F, H * W) + 0.5
    j = j.reshape(1, 1, H * W).expand(B, F, H * W) + 0.5

    fx, fy, cx, cy = K.chunk(4, dim=-1)

    x = (i - cx) / fx
    y = (j - cy) / fy
    z = torch.ones_like(x)

    dirs = torch.stack([x, y, z], dim=-1)
    dirs = dirs / dirs.norm(dim=-1, keepdim=True)

    rays_d = dirs @ c2w[..., :3, :3].transpose(-1, -2)
    rays_o = c2w[..., :3, 3][:, :, None].expand_as(rays_d)

    moments = torch.cross(rays_o, rays_d, dim=-1)
    plucker = torch.cat([moments, rays_d], dim=-1)

    return plucker.reshape(B, F, H, W, 6)
```

```python
source_hw = images.shape[-2:]      # VGGT processed image H, W
target_hw = (256, 384)             # CameraCtrl default H, W

c2w = w2c_3x4_to_c2w_4x4(extrinsic)
c2w = make_relative_c2w(c2w)

K = intrinsics_to_k4(intrinsic, source_hw, target_hw)

plucker = plucker_from_cameras(K, c2w, *target_hw)       # [B, F, H, W, 6]
plucker_bfchw = plucker.permute(0, 1, 4, 2, 3).contiguous()  # [B, F, 6, H, W]
plucker_cameractrl = plucker_bfchw.permute(0, 2, 1, 3, 4).contiguous()  # [B, 6, F, H, W]
```
